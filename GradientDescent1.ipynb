```python
from sklearn.datasets import make_regression
import numpy as np
from matplotlib import pyplot as plt
```

```python
X,y=make_regression(n_samples=4,n_features=1,noise=80,n_targets=1,random_state=13)
print(X)
print(y)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color="red")
```

```python
#ASSUME m=78.35 and b=0
y_pred=((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color="red",label="OLS")
plt.plot(X,y_pred,color="green",label="b=0")
plt.legend()
```

```python
m=78.35
b=100
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
lr=0.1
step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1=(m*X+b).reshape(4)
```

```python
plt.scatter(X,y)
#plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python

# Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = lr * loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=(m*X+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred2,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
```

```python
b=b-step_size
b
```
