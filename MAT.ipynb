**MATPLOTLIB**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(y,z)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
# ADDING TITLE TO AXIS
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sine wave")
plt.show()
```

```python
# PLOT IN DOTS
plt.plot(x,y,"o")
plt.show()
```

```python
plt.plot(x,y,"rx")
plt.show()
```

```python
# 2 GRAPHS in one
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

Bar plot


```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
langs=["c","c++","java","python","php"]
students=[23,17,35,29,12]
ax.bar(langs,students)
plt.show()
```
