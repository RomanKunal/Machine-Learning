```python
#loading the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
import sklearn.datasets
from sklearn.svm import SVC
```

```python
#load the dataset
cancer_dataset=sklearn.datasets.load_breast_cancer()
print(cancer_dataset)
```

```python
#loading the dataset into dataframe
data_frame=pd.DataFrame(cancer_dataset.data,columns=cancer_dataset.feature_names)
data_frame.head()
```

```python
#adding the target column
data_frame['label']=cancer_dataset.target
data_frame.head()
```

```python
data=data_frame.copy()
```

```python
data['label'].value_counts()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
#separting features
x=data.drop(columns='label',axis=1)
y=data['label']
```

```python
X=np.array(x)
Y=np.array(y)
```

#grid search cv
is used to determining the best parameter for our model

```python
#loading the svc
svc=SVC()
```

```python
#hyperparamter
parameters={'kernel':['linear','poly','sigmoid','rbf'],'C':[1,10,15,20]}
```

```python
#grid search cv
grid_search=GridSearchCV(svc,parameters,cv=5)
```

```python
#fitting the data into model
grid_search.fit(X,Y)
```

```python
grid_search.cv_results_
```

```python
#finding the best parameter
best=grid_search.best_params_
print(best)
```

```python
#loading the dataset into dataframe
res=pd.DataFrame(grid_search.cv_results_)
res.head()
```

```python
#highest accuracy
highest=grid_search.best_score_
print(highest)
```

```python
#average score
grid_search_result=res[['param_C','param_kernel','mean_test_score']]
```
