# Bivariate Analysis

```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
train=pd.read_csv("/content/train.csv")
train
```

```python
flights=sns.load_dataset("flights")
flights.head()
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

# ScatterPlot
(numerical-numerical)

```python
sns.scatterplot(data=tips, x="total_bill", y="tip",hue='sex',style='smoker')
```

# BAR PLOT
(numerical-categorical)

```python
sns.barplot(data=train,x='Pclass',y='Age',hue='Sex')
```

```python
sns.boxplot(data=train,x='Age',y='Sex',hue='Survived')
```

# Dist Plot
(Numerical-categorical)

```python
sns.distplot(train[train['Survived']==0]['Age'],hist=False)
sns.distplot(train[train['Survived']==1]['Age'],hist=False)
```

```python
sns.heatmap(pd.crosstab(train['Pclass'],train['Survived']))
```
