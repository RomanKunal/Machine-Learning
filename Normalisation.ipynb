```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/wine_data.csv")
df
```

```python
df1=pd.read_csv("/content/wine_data.csv",usecols=[0,1,2],header=None)
df1.columns=['Class label', 'Alcohol', 'Malic acid']
df1
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df1.drop('Class label',axis=1),
                                               df1['Class label'],
                                               test_size=0.3,
                                               random_state=0)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train)

X_scaled=scaler.transform(X_train)
X_testscaled=scaler.transform(X_test)

```

```python
X_scaled=pd.DataFrame(X_scaled,columns=X_train.columns)
X_testscaled=pd.DataFrame(X_testscaled,columns=X_test.columns)
```
