# Working on the data and time handling


```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
time=pd.read_csv("/content/messages.csv")
date=pd.read_csv("/content/orders.csv")
```

**If we want to manage the date and time so we need to must convert the datatype into dateandtime using pandas**


```python
date.info()
```

```python
time.info()
```

```python
date['date']=pd.to_datetime(date['date'])
date.info()
```

```python
date['data_year']=date['date'].dt.year
date['data_month']=date['date'].dt.month
date['data_day']=date['date'].dt.day
date.head()
```

```python
date['weekday']=date['date'].dt.dayofweek
date.head()
```

```python
date['weekname']=date['date'].dt.day_name()
date.head()
```

```python
date['weekend']=np.where(date['weekname'].isin(['Saturday','Sunday']),1,0)
date.head()
```

```python
date['date_week'] = date['date'].dt.week

date.head()
```

```python
time['date']=pd.to_datetime(time['date'])
time.info()
```

```python
date['quarter'] = date['date'].dt.quarter
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

time.head()
```

```python
time['time']=time['date'].dt.time
time.head()
```
