# HANDLING IMBALANCED DATASET
1. a dataset with an unequal distribution

```python
cc=pd.read_csv("/content/credit_data.csv")
cc.head()
```

```python
cc.isnull().sum()
```

```python
cc["Class"].value_counts()
```

```python
# seperate the fraud and legit trans as this is highly imbalanced dataset
legit=cc[cc.Class==0]
fraud=cc[cc.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# BUILD A SAMPLE DATASET CONTAINING SIMILAR TRANS OD LEGIT AND FRAUD
legit_sample=legit.sample(n=492)
print(legit_sample.shape)
```

```python
# concat
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```
