```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data=pd.read_csv('/content/mail_data.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.tail()
```

```python
mail_data=data.where((pd.notnull(data)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

```python
mail_data.loc[mail_data['Category']=='spam','Category',]=0
mail_data.loc[mail_data['Category']=='ham','Category',]=1
```

```python
mail_data.head()
```

```python
from re import X
# separting text and labels
X=mail_data['Message']
Y=mail_data['Category']
```

```python
print(X.shape)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# CONVERT TEXT DATA INTO NUMERICALS to use in logisitic regression
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)

Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
# IMPLEMENT LOGISITIC REGRESSION
model=LogisticRegression()
model.fit(X_train_features,Y_train)
```

```python
# EVALATION
prediction_on_training_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
print('Accuracy on training data:',accuracy_on_training_data*100)
```

```python
# PREDICTING ON TESTING DATA
prediction_on_testing_data=model.predict(X_test_features)
accuracy_on_testing_data=accuracy_score(Y_test,prediction_on_testing_data)
print('Accuracy on testing data:',accuracy_on_testing_data*100)
```
