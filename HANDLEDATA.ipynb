**method to handle missing values**
1. IMPUTATION
2. DROPPING


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/Placement_Dataset.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

CENTRAL TENDENCY
1. MEAN
2. MEDIAN
3. MODE


```python
fig,ax=plt.subplots(figsize=(15,5))
sns.distplot(data['salary'])
```

```python
# AS WE HAVE OUTLIER IN DATA SO USE MEDIAN AND MODE
# MEAN SHOULD BE USED IN SKEW DATA
data['salary'].fillna(data['salary'].median(),inplace=True)
```

```python
data.isnull().sum()
```

```python
# DROPPING METHOD
salary_dataset=pd.read_csv('/content/Placement_Dataset.csv')
salary_dataset.head()
```

```python
salary_dataset.isnull().sum()
```

```python
# dropping
salary_dataset.dropna(inplace=True)
```
