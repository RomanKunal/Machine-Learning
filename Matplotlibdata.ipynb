**SUBPLOT**
  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("/content/Data/batter.csv")
data
```

```python
newdata=data.head(20)
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1)
ax[0].scatter(newdata["runs"],newdata["avg"],color="yellow")
ax[1].scatter(newdata["avg"],newdata["strike_rate"],color="Red")
ax[0].set_title("Runs Vs Avg")
ax[1].set_title("Avg vs StrikeRate")
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2)
ax[0,0].scatter(newdata["runs"],newdata["avg"],color="yellow")
ax[0,1].scatter(newdata["avg"],newdata["strike_rate"],color="Red")
ax[0,0].set_title("Runs Vs Avg")
ax[0,1].set_title("Avg vs StrikeRate")
ax[1,0].hist(newdata["runs"])
ax[1,1].hist(newdata["strike_rate"])
```

**3D plots**

```python
newdata
fig=plt.figure()
ax=plt.subplot(projection="3d")
ax.scatter3D(newdata["runs"],newdata["avg"],newdata["strike_rate"])
ax.set_title("runs VS avg VS strike rate")
ax.set_xlabel("Runs")
ax.set_ylabel("Average")
ax.set_zlabel("Strike Rate")
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
plt.colorbar(p)
```
