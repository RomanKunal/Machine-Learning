# LABEL ENCODING
1> converting the labels into numeric form


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

LABELING ENCODING OF BREAST CANCEER

```python
cancerdata=pd.read_csv('/content/breast_cancer_data.csv')
cancerdata.head()
```

```python
cancerdata.isnull().sum()
```

```python
cancerdata['diagnosis'].value_counts()
```

```python
#load the label encoder
label=LabelEncoder()
```

```python
labels=label.fit_transform(cancerdata.diagnosis)
```

```python
print(labels)
```

```python
# appending the labels to the datset
cancerdata['target']=labels
```

```python
cancerdata.head()
```

# LABEL ENCODING OF IRIS DATASET


```python
iris=pd.read_csv('/content/iris_data.csv')
iris.head()
```

```python
iris['Species'].value_counts()
```
